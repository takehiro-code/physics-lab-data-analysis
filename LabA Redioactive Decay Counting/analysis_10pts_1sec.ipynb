# Radioactive Counting Experiment: 10 points 1 second

## Load required libraries

```python
# For scientific computing
import numpy as np

# For plotting
import matplotlib.pyplot as plt
```

## Load data for case without paper and with paper

```python
t1, R1 = np.genfromtxt('data/2020-09-29-10pts1sec.csv', delimiter=',', unpack=True)
print("t1 = ", t1)
print("R1 = ", R1)
print()

t2, R2 = np.genfromtxt('data/2020-09-29-10pts1sec_withPaper.csv', delimiter=',', unpack=True)
print("t2 = ", t2)
print("R2 = ", R2)
```

```python
# now check that it worked
mean1 = np.mean(R1)
stdev1 = np.std(R1, ddof=1)
print('Without paper: mean : ' + str(np.around(mean1, decimals=2)) + 
      ', standard devation: ' + str(np.around(stdev1, decimals=2)))
print()

# now check that it worked
mean2 = np.mean(R2)
stdev2 = np.std(R2, ddof=1)
print('With paper: mean: ' + str(np.around(mean2, decimals=2)) + 
      ', standard devation: ' + str(np.around(stdev2, decimals=2)))
```

```python
# and plot the data to have a look
plt.plot(t1,R1, "-o")
plt.title('10 pts 1 sec: R vs t plot WITHOUT paper (' + r'$\mu=$' + str(mean1)[:5] + ', ' + r'$\sigma=$' + str(stdev1)[:5] + ')')
plt.xlabel('t (s)')
plt.ylabel('R (Counts)')
plt.grid(which='major', axis='both')
plt.show()

# and plot the data to have a look
plt.plot(t2,R2, "-o")
plt.title('10 pts 1 sec: R vs t plot WITH paper (' + r'$\mu=$' + str(mean2)[:5] + ', ' + r'$\sigma=$' + str(stdev2)[:5] + ')')
plt.xlabel('t (s)')
plt.ylabel('R (Counts)')
plt.grid(which='major', axis='both')
plt.show()
```

## Plot the histogram

Alternatively, use matplotlib.pyplot.hist() with matplotlib.pyplot.errorbar(), instead of using matplotlib.pyplot.bar(). matplotlib.pyplot.hist() both bins the data and plots it. Adding the errorbars using matplotlib.pyplot.errorbar() provides a more universal way to plot error bars.

```python
# without paper
# plot histogram
offset = 2
bins = np.arange(np.min(R1)-offset, np.max(R1)+offset) - 0.5
counts, bins, patches = plt.hist(R1, bins=bins, density=True, rwidth=0.75)
plt.title('10 pts 1 sec: R vs t of Histogram WITHOUT paper')
plt.xlabel('Bin')
plt.ylabel('Normalized Probability')
plt.show()

# add the error bars
counts, bins, patches = plt.hist(R1, bins=bins, density=True, rwidth=0.75)
counts_err = np.sqrt(counts) / np.sqrt(t1.shape * np.diff(bins)) 

bcenter = (bins[:-1] + bins[1:]) / 2
plt.errorbar(bcenter, counts, yerr=counts_err, fmt='ko', markersize=0, capsize=4)

# Now label your graph
plt.title('10 pts 1 sec: R vs t of Histogram WITHOUT paper')
plt.xlabel('Bin')
plt.ylabel('Normalized Probability')
plt.show()
```

```python
# with Paper

# plot histogram
offset = 2
bins = np.arange(np.min(R2)-offset, np.max(R2)+offset) - 0.5
counts, bins, patches = plt.hist(R2, bins=bins, density=True, rwidth=0.75)
plt.title('10 pts 1 sec: R vs t of Histogram WITH paper')
plt.xlabel('Bin')
plt.ylabel('Normalized Probability')
plt.show()

# add the error bars
counts, bins, patches = plt.hist(R2, bins=bins, density=True, rwidth=0.75)
counts_err = np.sqrt(counts) / np.sqrt(t2.shape * np.diff(bins)) 

bcenter = (bins[:-1] + bins[1:]) / 2
plt.errorbar(bcenter, counts, yerr=counts_err, fmt='ko', markersize=0, capsize=4)

# Now label your graph
plt.title('10 pts 1 sec: R vs t of Histogram WITH paper')
plt.xlabel('Bin')
plt.ylabel('Normalized Probability')
plt.show()
```
